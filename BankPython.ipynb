### Project Description:

   The customer churn dataset is a collection of customer data that focuses on predicting customer churn, which refers to the tendency of
customers to stop using a companys products or services. The dataset contains various features that describe each customer, such as their 
credit score, country, gender, age, tenure, balance, number of products, credit card status, active membership, estimated salary, and churn status. 
The churn status indicates whether a customer has churned or not. The dataset is used to analyze and understand factors that contribute to customer
churn and to build predictive models to identify customers at risk of churning. The goal is to develop strategies and interventions to reduce churn 
and improve customer retention.

### Data Description:

- The dataset contains 12 columns and 10,000 rows.
-  Key features include:
      - Numerical columns: credit_score, age, tenure, balance, estimated_salary, etc.
    - 	Categorical columns: country, gender.
    -	Target variable: churn (indicating if a customer has churned).
    - All columns have 10,000 non-null entries.

```python
plt.figure(figsize=(10,10))
for i in numerical_columns:
    plt.subplot(3,3,numerical_columns.index(i)+1)
    sns.kdeplot(x=df[i],data=df,hue=df['churn'])
```

```python
#Assuming 'df' is your DataFrame, and 'numerical_column' contains the list of numerical columns.

plt. figure(figsize=(12, 10)) # Adjust the figure size

# Loop through numerical columns and create box plots
for idx, col in enumerate (numerical_columns) :
    plt.subplot(3, 3, idx + 1)# Create a grid layout
    sns.boxplot(x='churn', y=col, data=df,hue=df['churn'],palette='coolwarm')
    plt.title(f'Boxplot of {col} by Churn') 
    plt.xlabel('Churn')
    plt.ylabel(col)
    
plt.tight_layout() # Adjust spacing between plots 
plt.show()
```

```python
gender_churn = df.groupby('gender')['churn'].value_counts(normalize=True).unstack()
gender_churn.plot(kind='bar', stacked=True, figsize=(4, 4), colormap='coolwarm')
plt.title('Churn Proportion by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion of Churn')
plt.show()
```

```python
# Proportion of churn by country (example)
country_churn = df.groupby('country')['churn'].value_counts(normalize=True).unstack()
country_churn.plot(kind='bar', stacked=True, figsize=(4, 4), colormap='coolwarm')
plt.title('Churn Proportion by Country')
plt.xlabel('Country')
plt.ylabel('Proportion of Churn')
plt.show()
```

```python
# 6. Correlation heatmap for numerical features
correlation_matrix = df[numerical_columns + ['churn']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

### CONCLUSION:

 - Customer churn is influenced significantly by age, active member status, and balance.
Older customers and those with high account balances but low engagement (inactive membership) are more likely to leave the bank.
